```python
def find_by_name(name, list_of_albums):
    name = list(filter(lambda album: album['album'] == name, list_of_albums))
    if name: 
        return name[0]
    else:
        return None
```

```python
def find_by_rank(rank, list_of_albums):
    result = list(filter(lambda album: album['number'] == rank, list_of_albums))
    if result: 
        return result[0]
    else:
        return None
```

```python
def find_by_year(year, list_of_albums):
    results = list(filter(lambda album: album['year'] == year, list_of_albums))
    return results
```

```python
def most_popular_word(list_of_albums):
    try:
        albumlist = list(map(lambda album: album['album'].split(), list_of_albums))
    except:
        albumlist = list(map(lambda album: album['name'].split(), list_of_albums))
    words = []
    for sublist in albumlist:
        for item in sublist:
            words.append(item)
    return max(words, key=words.count)
```

```python
print(most_popular_word(albums))
```

```python
songlist = []
for item in lines:
    songdict = {}
    song = item.split("\t")
    songdict['number'] = int(song[0])
    songdict['name'] = song[1]
    songdict['artist'] = song[2]
    songdict['year'] = int(song[3])
    songlist.append(songdict)
print(songlist)
```

```python
import json

file = open('track_data.json', 'r')
tracks_data = json.load(file)

print(tracks_data)
```

```python
def albumwithmosttopsongs():
    songs = list(map(lambda song: song['name'], songlist))


    album_tracks = {}

    for album in tracks_data:
        album_tracks[album['album']] = 0
        for track in album['tracks']:
            if track in songs:
                album_tracks[album['album']] += 1
    topalbum = max(album_tracks, key=album_tracks.get)
    for album in tracks_data: 
        if album['album'] == topalbum:
            topartist = album['artist']
  
    return topalbum + " by " + topartist 
```

```python
albumwithmosttopsongs()
```

```python
def albumswithtopsongs():
    songs = list(map(lambda song: song['name'], songlist))
    topalbums = []
    for album in tracks_data:
        for track in album['tracks']:
            if track in songs:
                topalbums.append(album['album'])
    return list(set(topalbums))
```

```python
def songsthatareontopalbums():
    songs = list(map(lambda song: song['name'], songlist))
    topsongs = []
    for album in tracks_data:
        for track in album['tracks']:
            if track in songs:
                topsongs.append(track)
    return list(set(topsongs))
```

```python
from collections import Counter
def topoverallartist():
    artistlistalbums = list(map(lambda album: album['artist'], albums))
    artistlisttracks = list(map(lambda song: song['artist'], songlist))
    topoverall = dict(Counter(artistlistalbums) + Counter(artistlisttracks))
    return max(topoverall, key=topoverall.get)
```
